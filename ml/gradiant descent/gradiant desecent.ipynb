```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
def gradiant(x,y):
    m=b=0
    rate=0.08
    n=len(x)
    iteration=100
    for i in range(iteration):
        y_predict=m*x+b
        plt.plot(x,y_predict,color='red')
        md=-(2/n)*sum(x*(y-y_predict))
        yd=-(2/n)*sum(y-y_predict)
        m=m-rate*md
        b=b-rate*yd
        cost=-(1/n)*sum(val**2 for val in (y-y_predict))
        print(f"cost{cost},b{b},m{m},iteration {i}")
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
gradiant(x,y)
```
